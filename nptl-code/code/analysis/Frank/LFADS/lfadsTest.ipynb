```python
import numpy
estRates = numpy.asarray(lfResult.get('/output_dist_params'))
trueRates = numpy.asarray(truth.get('/valid_truth'))
spikeRasters = numpy.asarray(truth.get('/valid_data'))
conversionFactor = numpy.asarray(truth.get('/conversion_factor'))
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
for x in range(0,50):
    plt.subplot(8,8,x+1)
    plt.plot(estRates[50,:,x]*conversionFactor,'r')
    plt.plot(trueRates[50,:,x],'b')
    
plt.figure(2)
for x in range(0,50):
    plt.subplot(8,8,x+1)
    plt.plot(estRates[50,:,x]*conversionFactor,'r')
    plt.plot(spikeRasters[50,:,x],'b')
    
plt.show()
```
